```python
import numpy as np 
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
df20=pd.read_csv(r"Ratings\EAplayers_20.csv")# EA player rating csv file downloaded from kaggle 
df19=pd.read_csv(r"Ratings\EAplayers_19.csv") 
df18=pd.read_csv(r"Ratings\EAplayers_18.csv") 
df17=pd.read_csv(r"Ratings\EAplayers_17.csv") 
df16=pd.read_csv(r"Ratings\EAplayers_16.csv") 
df15=pd.read_csv(r"Ratings\EAplayers_15.csv") 
```

```python
fifa20=df20
fifa20=fifa20[['short_name','club','overall','team_position','pace','shooting','passing','dribbling','defending','physic','gk_diving']]
drop_subs  = fifa20[ fifa20['team_position'] == 'SUB' ] # selecting subs from team position column 
drop_reserves  = fifa20[ fifa20['team_position'] == 'RES' ] 
fifa20 = fifa20.drop(drop_subs.index, axis=0) #dropping subs from the fifa20 dataframe
fifa20 = fifa20.drop(drop_reserves.index, axis=0)
team = fifa20.groupby('club')#group by teams
pl_team=['Arsenal','Aston Villa','Bournemouth','Brighton','Burnley','Chelsea','Crystal Palace','Everton','Leicester','Liverpool','Man City','Man United','Newcastle','Norwich','Stoke','Southampton','Tottenham','Watford','West Ham','Wolves']
pl_club=[0,1,2,3,4,6,7,8,12,13,14,15,17,18,21,20,24,25,27,28]
pl_teams=[]

premier_league=pl_teams

for x in range (len(pl_team)):
    pl_teams.append(team.get_group(pl_team[x])) #append the team to pl_teams

#FEATURE OVERALL MEDIAN TEAM RATING
overall=[]
for x in range (len(pl_team)):
    pl_team[x]=premier_league[x]
    overall.append(pl_team[x]['overall'].mean())

#EXTRA FEATURES BASED ON DEFENSE OR ATTACK 
physical=[]
defending=[]
pace=[]
shooting=[]
passing=[]
dribbling=[]
gk=[]
for x in range(len(pl_team)):# store ratings based on positon
    oi=pl_team[x].loc[pl_team[x]['team_position'].str.contains('B|D')]
    defending.append(oi['defending'].mean())
    physical.append(oi['physic'].mean())

for x in range(len(pl_team)):
    oi=pl_team[x].loc[pl_team[x]['team_position'].str.contains('GK')]
    gk.append(oi['gk_diving'].mean())

    
for x in range(len(pl_team)):
    oi=pl_team[x].loc[pl_team[x]['team_position'].str.contains('B|D|GK')]
    pace.append(pl_team[x]['pace'].mean())
    shooting.append(pl_team[x]['shooting'].mean())
    passing.append(pl_team[x]['passing'].mean())
    dribbling.append(pl_team[x]['dribbling'].mean())

EA_20 = pd.DataFrame({'Release_year':2019, 'Club':pl_club, 'Overall':overall,'Physical':physical,'Defending':defending,'Pace':pace,'Shooting':shooting,'Passing':passing,'Dribbling':dribbling,'GK_diving':gk})#create dataframe with above columns and adding release year
```
